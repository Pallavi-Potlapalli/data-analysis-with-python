```python
import pandas as pd
mydf3=pd.read_csv("/content/WA_Fn-UseC_-HR-Employee-Attrition.csv")
mydf3
```

```python
#trees
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
DistanceFromHome=int(input("enter DistanceFromHome"))
DailyRate=int(input("enter DailyRate"))
WorkLifeBalance=int(input("enter WorkLifeBalance"))
JobSatisfaction=int(input("enter JobSatisfaction	"))
features_col=['DistanceFromHome','DailyRate','WorkLifeBalance','JobSatisfaction']
features=mydf3[features_col]
target=mydf3.Attrition
DT2=DecisionTreeClassifier()
model=DT2.fit(features,target)
a=model.predict([[DistanceFromHome,DailyRate,WorkLifeBalance,JobSatisfaction]])
if a=="Yes":
  print("leave the company")
else:
  print("not leave the company")
```

```python
from sklearn.tree import export_graphviz,plot_tree
plot_tree(DT2)
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
DistanceFromHome=int(input("enter DistanceFromHome"))
DailyRate=int(input("enter DailyRate"))
WorkLifeBalance=int(input("enter WorkLifeBalance"))
JobSatisfaction=int(input("enter JobSatisfaction	"))
features_col=['DistanceFromHome','DailyRate','WorkLifeBalance','JobSatisfaction']
features=mydf3[features_col]
target=mydf3.Attrition
RF2=RandomForestClassifier()
model=RF2.fit(features,target)
n_estimators=10
a=model.predict([[DistanceFromHome,DailyRate,WorkLifeBalance,JobSatisfaction]])
if a=="Yes":
  print("leave the company")
else:
  print("not leave the company")
```

```python
import os
output_dir ="tree_visualizations_on_HR_employee_attrition"
os.makedirs(output_dir,exist_ok=True)
```

```python
for i, tree in enumerate(model.estimators_):
  tree_dot_file=os.path.join(output_dir,f"tree_{i}.dot")
  tree_png_file=os.path.join(output_dir,f"tree_{i}.png")
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file=tree_dot_file, feature_names=['DistanceFromHome','DailyRate','WorkLifeBalance','JobSatisfaction'],
                    class_names=[str(cls) for cls in model.classes_], filled=True,rounded=True)
```

```python
command=f"dot -Tpng {tree_dot_file} -o {tree_png_file}"
os.system(command)
print(f"Tree {i} visualization saved to {tree_png_file}")
```

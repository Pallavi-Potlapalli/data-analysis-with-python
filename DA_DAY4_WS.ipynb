```python
u="https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
```

```python
p=requests.get(u)
```

```python
c=BeautifulSoup(p.text,'html')
print(c)
```

```python
t1=c.find('table')
print(t1)
```

```python
h=t1.find_all('th')
print(h)
```

```python
headings = [i.text for i in h]
print(headings)
```

```python
import pandas as pd
```

```python
mydf = pd.DataFrame(columns=headings)
print(mydf)
```

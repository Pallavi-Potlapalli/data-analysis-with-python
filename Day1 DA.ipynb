```python
#Get the  salary from the user if the salary is more than or equal to 50000 apply 10% tax deduction if the salary is less then 50000 apply 5% tax if the salary is more than 70000 then apply 12% tax if the salary exceeds 1lakh print your are promoted to next level
salary = int(input("Enter your salary: "))
if salary >= 100000:
    print("Congratulations! You are promoted to the next level.")
    tax_percentage = 10
elif salary >= 70000:
    tax_percentage = 12
elif salary >= 50000:
    tax_percentage = 10
else:
    tax_percentage = 5
tax_amount = (tax_percentage / 100) * salary
net_salary = salary - tax_amount
print("Your salary before tax: {salary}")
print("Tax deduction at {tax_percentage}%: {tax_amount}")
print("Your net salary after tax: {net_salary}")
```

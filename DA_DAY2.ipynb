```python
#there are 5 subjects psychology,nano tecnology,medicine,oncology,anthropology. two students applied to do masters in any of these 5 fields, with a line plot show in which field they can excel. the cutoff for each field is 70.
#[65,45,73,71,57]
#[43,67,68,53,85]
```

```python
sub = ["Psy","Nano","Med","Onc","Anthro"]
s1 = [67,45,73,71,57]
s2 = [43,67,68,53,85]
cut = [70,70,70,70,70]
#student 1
#plt.plottype(x,y,label,color,marker,linewidth,linestyle)
plt.scatter(sub,s1,label="student 1",color ="black",marker="p")
#student 2
plt.scatter(sub,s2,label="student 2",color ="blue",marker='H')
#cutoff
plt.plot(sub,cut,label ="cutoff",color ="green",linestyle=":",linewidth=5)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
```

#an india vs australia is on live.india has given 298 as target for australia. the runs for every 5 overs is given as follows
#[25,51,84,131,160,189,220,250,267,297]
#australia has completed
#[15,41,94,110,151] plot overs vs runs for both the countries with india as blue line and yellow line as australia with required labels and lengends the marker for india is flag and the marker for australia is + take overs from arange function create arrays for every runs linewidth=2

```python
import matplotlib.pyplot as plt
import numpy as np
overs_india = np.arange(5, 51, 5)
runs_india = [25, 51, 84, 131, 160, 189, 220, 250, 267, 297]
overs_aus = np.arange(5, 26, 5)
runs_aus = [15, 41, 94, 110, 151]
plt.plot(overs_india, runs_india, 'b', label='India', marker='>', linewidth=3)
plt.plot(overs_aus, runs_aus, 'y-', label='Australia', marker='+', linewidth=3)
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.title('India vs Australia Live Match')
plt.legend()
plt.xticks(np.arange(0, 51, 5))
plt.show()
```
